```python
import random
tao = []

def read_file():
    with open('/home/jhilene/markov/taotc.txt', 'r') as f:
        for line in f:
            line = line.replace(".", " .")
            for word in line.lower().strip().split():
            #empty.append(line.split())
                tao.append(word)
        print tao
        
read_file()
```

```python
master_dict = {}
total = 0
for i in range(len(tao)-1):
    j = i + 1
    t_w = tao[i]
    j_w = tao[j]
    if t_w not in master_dict:
        master_dict[t_w] = {'_total':0}
    w_d = master_dict[t_w]
         
    if j_w not in w_d: 
       w_d[j_w] = 0
    if j_w in w_d:
        w_d[j_w] += 1
        w_d['_total'] += 1
   
print master_dict
    
```

```python
def weighted_choice(choices):
   total = sum(w for c, w in choices.iteritems() if c != '_total')
   r = random.uniform(0, total)
   #print choices
   #print r
   upto = 0
   for c, w in choices.iteritems():
      if c == '_total':
        continue
      if upto + w >= r:
         #print c
         return c
      upto += w
   assert False, "Shouldn't get here"

# [ a a a b b c c c c c ]

# { 1: a, 4: b, 6: c }

# [ a * * b * c * * * * ]


for i in range(20):
   weighted_choice(master_dict['compassion'])
```
